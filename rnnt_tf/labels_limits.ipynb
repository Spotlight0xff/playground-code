```python
def test_limit_instance(n, n_time, n_target): 
    phase_a_end = n_time-1 if n_target > n_time else n_target-1
    phase_b_end = n_target-1 if n_target > n_time else n_time-1
    phase_c_end = n_target + n_time - 2
    
    print(f"[n={n}] phases: A={phase_a_end}; B={phase_b_end}; C={phase_c_end}")
    if n <= phase_a_end:
        phase = "A"
        print(f"[n={n}] phase (a),        ", end="")
        limits = (0, n)
    elif n <= phase_b_end and n_time >= n_target:
        phase = "B"
        print(f"[n={n}] phase (b), T > U, ", end="")
        limits = (0, n_target-1)
    elif n <= phase_b_end and n_target > n_time:
        phase = "B"
        print(f"[n={n}] phase (b), U > T, ", end="")
        limits = (n-n_time+1, n)
    # phase (c)
    elif n > phase_b_end and n_time > n_target:
        phase = "C"
        print(f"[n={n}] phase (c), T > U, ", end="")
        limits = (n-n_time+1, n_target-1)
    elif n > phase_b_end and n_target >= n_time:
        phase = "C"
        print(f"[n={n}] phase (c), U > T, ", end="")
        limits = (n-n_time+1, n_target-1)
    else:
        assert False, f"U={n_target}, T={n_time}, n={n}"
        
    print(f"limits", limits, "\n")
    num_labels = limits[1] - limits[0]
    if phase == "A":
        assert num_labels == n
    elif phase == "B":
        phase_b_num_labels = min(n_target, n_time) - 1
        # assert num_labels == phase_b_num_labels
    elif phase == "C":
        phase_c_num_labels = min(n_target, n_time)
        #assert limits[1] - limits[0] == 
    
    assert limits[1] - limits[0] > 0
    assert limits[1] - limits[0] <= n_target
    assert limits[1] < n_target + 1
    
def test_limits(n_time, n_target):
    print(f"test limits T={n_time}, U={n_target}")
    for n in range(1, n_time+n_target-2):
        test_limit_instance(n, n_time, n_target)
    print()


test_limits(n_target=8, n_time=3)
test_limits(n_target=5, n_time=5)
test_limits(n_target=3, n_time=8)
test_limits(n_target=5, n_time=4)
```
